# Map vote

```python
from map_creator import create_map
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
year = 2017
```

```python
election_data = pd.read_parquet(
    f"../data/elections/presidentiel/{year}/pres{year}_csv/pres{year}comm.parquet"
)
```

```python
def style_function(feature):
    # Get the commune code and result
    codecommune = feature["properties"]["code"]
    result = election_data.loc[election_data["codecommune"] == codecommune, "pvoteTG"]

    # Ensure result is not empty and has only one value
    if len(result) > 0:
        result_value = result.iloc[0]
    else:
        result_value = 0  # Default value if no result

    # Normalize the result to be between 0 and 1, if needed
    result_value = np.clip(result_value, 0, 1)  # Ensuring it's within [0, 1]

    # Generate a color from the "viridis" color map
    color = plt.cm.viridis(
        result_value
    )  # Map the result value to a color in the color map

    # Convert the RGBA value (0-1) to a hex color
    color_hex = matplotlib.colors.to_hex(color)

    return {
        "fillColor": color_hex,  # Apply the color to the fill
        "color": "black",  # Border color
        "fillOpacity": 0.7,  # Fill opacity
        "weight": 0.3,  # Border weight
    }
```

```python
create_map(
    "Vote à gauche_" + str(year),
    "map_created/france_vote_gauche_commune_" + str(year),
    style_function=style_function,
    legend=True,
)
```
